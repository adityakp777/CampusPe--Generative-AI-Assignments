```python
# Simple ATM Simulator

try:
    # Initial balance
    account_balance = 10000
    minimum_balance_required = 500

    while True:
        # Display menu
        print("\n+------------------------------------------------------+")
        print("|                 ATM SIMULATOR                      |")
        print("+------------------------------------------------------+")
        print("| 1. Check Balance                                   |")
        print("| 2. Deposit Money                                   |")
        print("| 3. Withdraw Money                                  |")
        print("| 4. Exit                                            |")
        print("+------------------------------------------------------+")

        user_choice = input("Enter choice (1-4): ")

        # Check Balance
        if user_choice == "1":
            print("\n+------------------------------------------------------+")
            print("| Current Balance: ₹{:.2f}".format(account_balance).ljust(53) + "|")
            print("+------------------------------------------------------+")

        # Deposit
        elif user_choice == "2":
            deposit_amount = float(input("Enter amount to deposit: "))

            if deposit_amount <= 0:
                print("Deposit amount must be greater than zero.")
            else:
                account_balance += deposit_amount
                print("Deposit successful!")
                print(f"Updated Balance: ₹{account_balance:.2f}")

        # Withdraw
        elif user_choice == "3":
            withdrawal_amount = float(input("Enter amount to withdraw: "))

            if withdrawal_amount <= 0:
                print("Withdrawal amount must be greater than zero.")

            elif withdrawal_amount > account_balance:
                print("Insufficient balance.")

            elif account_balance - withdrawal_amount < minimum_balance_required:
                print("Cannot withdraw. Minimum balance of ₹500 must remain.")

            else:
                account_balance -= withdrawal_amount
                print("Withdrawal successful!")
                print(f"Updated Balance: ₹{account_balance:.2f}")

        # Exit
        elif user_choice == "4":
            print("Thank you for using ATM Simulator.")
            break

        else:
            print("Invalid choice. Please select 1-4.")

except ValueError:
    print("Invalid input. Please enter numeric values where required.")
```

```python
# Number Pattern Printer

try:
    # Display menu
    print("\n+------------------------------------------------------+")
    print("|              NUMBER PATTERN PRINTER                |")
    print("+------------------------------------------------------+")
    print("| 1. Pattern 1 (Increasing numbers)                  |")
    print("| 2. Pattern 2 (Repeated row numbers)                |")
    print("| 3. Pattern 3 (Reverse decreasing pattern)          |")
    print("| 4. Pattern 4 (Pyramid pattern)                     |")
    print("+------------------------------------------------------+")

    pattern_choice = int(input("Enter pattern choice (1-4): "))
    height = int(input("Enter height of pattern: "))

    if height <= 0:
        print("Height must be greater than zero.")

    elif pattern_choice == 1:
        print("\nPattern 1:\n")
        for i in range(1, height + 1):
            for j in range(1, i + 1):
                print(j, end=" ")
            print()

    elif pattern_choice == 2:
        print("\nPattern 2:\n")
        for i in range(1, height + 1):
            for j in range(i):
                print(i, end=" ")
            print()

    elif pattern_choice == 3:
        print("\nPattern 3:\n")
        for i in range(height, 0, -1):
            for j in range(i, 0, -1):
                print(j, end=" ")
            print()

    elif pattern_choice == 4:
        print("\nPattern 4:\n")
        for i in range(1, height + 1):
            # Left increasing part
            for j in range(1, i + 1):
                print(j, end="")
            # Right decreasing part
            for j in range(i - 1, 0, -1):
                print(j, end="")
            print()

    else:
        print("Invalid pattern choice. Please select 1-4.")

except ValueError:
    print("Invalid input. Please enter numeric values only.")
```

```python
# Sum, Average, Median, Mode Calculator

from collections import Counter

try:
    total_numbers = int(input("How many numbers? "))

    if total_numbers <= 0:
        print("Number of inputs must be greater than zero.")
    else:
        numbers_list = []

        for i in range(1, total_numbers + 1):
            number_input = float(input(f"Enter number {i}: "))
            numbers_list.append(number_input)

        total_sum = sum(numbers_list)
        average_value = total_sum / total_numbers
        maximum_value = max(numbers_list)
        minimum_value = min(numbers_list)

        # Median
        sorted_numbers = sorted(numbers_list)
        if total_numbers % 2 == 0:
            mid1 = sorted_numbers[total_numbers // 2 - 1]
            mid2 = sorted_numbers[total_numbers // 2]
            median_value = (mid1 + mid2) / 2
        else:
            median_value = sorted_numbers[total_numbers // 2]

        # Mode
        frequency = Counter(numbers_list)
        max_count = max(frequency.values())
        mode_values = [num for num, count in frequency.items() if count == max_count]

        print(f"\nSum: {total_sum}")
        print(f"Average: {average_value:.2f}")
        print(f"Maximum: {maximum_value}")
        print(f"Minimum: {minimum_value}")
        print(f"Median: {median_value}")
        print(f"Mode: {mode_values}")

except ValueError:
    print("Invalid input. Please enter numeric values only.")
```

```python
# Factorial Calculator

try:
    number_input = int(input("Enter a number: "))

    # Handle negative numbers
    if number_input < 0:
        print("Factorial is not defined for negative numbers.")

    # Handle 0 case
    elif number_input == 0:
        print("0! = 1")

    else:
        factorial_result = 1
        calculation_steps = ""

        # Loop for factorial calculation
        for i in range(number_input, 0, -1):
            factorial_result *= i
            calculation_steps += str(i)
            if i != 1:
                calculation_steps += " x "

        # Display step-by-step result
        print(f"{number_input}! = {calculation_steps} = {factorial_result}")

except ValueError:
    print("Invalid input. Please enter an integer value only.")
```

```python
# Prime Number Checker

import math

try:
    # -------- Part 1: Single Number Prime Check --------
    number_input = int(input("Enter a number: "))

    if number_input < 2:
        print(f"{number_input} is NOT a prime number")

    elif number_input == 2:
        print("2 is a PRIME number")

    else:
        is_prime = True

        # Check divisibility up to sqrt(n)
        for i in range(2, int(math.sqrt(number_input)) + 1):
            if number_input % i == 0:
                is_prime = False
                break

        if is_prime:
            print(f"{number_input} is a PRIME number")
        else:
            print(f"{number_input} is NOT a prime number")

    # -------- Part 2: Prime Numbers in a Range --------
    start_range = int(input("\nEnter start range: "))
    end_range = int(input("Enter end range: "))

    if start_range > end_range:
        print("Invalid range. Start should be less than or equal to end.")
    else:
        prime_numbers_list = []

        for num in range(max(2, start_range), end_range + 1):
            is_prime = True
            for i in range(2, int(math.sqrt(num)) + 1):
                if num % i == 0:
                    is_prime = False
                    break
            if is_prime:
                prime_numbers_list.append(num)

        if prime_numbers_list:
            print("Prime numbers:", ", ".join(map(str, prime_numbers_list)))
        else:
            print("No prime numbers found in this range.")

except ValueError:
    print("Invalid input. Please enter integer values only.")
```

```python
import random

best_score = None

while True:
    print("\nSelect Difficulty Level:")
    print("1. Easy (1-50)")
    print("2. Medium (1-100)")
    print("3. Hard (1-1000)")

    difficulty_choice = input("Enter choice (1-3): ")

    if difficulty_choice == "1":
        lower, upper = 1, 50
    elif difficulty_choice == "2":
        lower, upper = 1, 100
    elif difficulty_choice == "3":
        lower, upper = 1, 1000
    else:
        print("Invalid choice.")
        continue

    target_number = random.randint(lower, upper)
    attempts_allowed = 7
    attempts_used = 0
    guessed_correctly = False

    print(f"\nGuess a number between {lower} and {upper}. You have 7 attempts.")

    while attempts_used < attempts_allowed:
        try:
            user_guess = int(input("Enter your guess: "))
            attempts_used += 1
            attempts_remaining = attempts_allowed - attempts_used

            if user_guess == target_number:
                print(f"Correct! You guessed in {attempts_used} attempts.")
                guessed_correctly = True

                if best_score is None or attempts_used < best_score:
                    best_score = attempts_used
                    print("New best score!")

                break

            elif user_guess < target_number:
                print("Too low!")
            else:
                print("Too high!")

            if abs(user_guess - target_number) <= 5:
                print("Hint: You are very close!")

            print(f"Attempts remaining: {attempts_remaining}")

        except ValueError:
            print("Enter a valid integer.")

    if not guessed_correctly:
        print(f"Game Over! The number was {target_number}.")

    if best_score is not None:
        print(f"Best score: {best_score}")

    play_again = input("Play again? (yes/no): ").lower()
    if play_again != "yes":
        break
```

```python
# Palindrome Checker

try:
    user_input = input("Enter word/number: ")

    if user_input.strip() == "":
        print("Input cannot be empty.")
    else:
        original_value = user_input
        processed_value = user_input.lower()

        reversed_value = processed_value[::-1]

        print(f"\nOriginal: {original_value}")
        print(f"Reversed: {original_value[::-1]}")

        print("\nStep-by-step comparison:")

        is_palindrome = True

        for i in range(len(processed_value) // 2):
            left_char = processed_value[i]
            right_char = processed_value[-(i + 1)]

            print(f"Comparing '{left_char}' with '{right_char}'")

            if left_char != right_char:
                is_palindrome = False
                break

        if is_palindrome:
            print("\nResult: PALINDROME")
        else:
            print("\nResult: NOT A PALINDROME")

except Exception:
    print("Unexpected error occurred.")
```

```python
# Text Analysis Functions

import string


# 1. Count Words
def count_words(text):
    words = text.split()
    return len(words)


# 2. Count Vowels
def count_vowels(text):
    vowels = "aeiou"
    count = 0
    for char in text.lower():
        if char in vowels:
            count += 1
    return count


# 3. Count Consonants
def count_consonants(text):
    vowels = "aeiou"
    count = 0
    for char in text.lower():
        if char.isalpha() and char not in vowels:
            count += 1
    return count


# 4. Reverse Text
def reverse_text(text):
    return text[::-1]


# 5. Check Palindrome
def is_palindrome(text):
    cleaned_text = "".join(char.lower() for char in text if char.isalnum())
    return cleaned_text == cleaned_text[::-1]


# 6. Remove Vowels
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return "".join(char for char in text if char not in vowels)


# 7. Word Frequency
def word_frequency(text):
    words = text.lower().split()
    frequency = {}
    for word in words:
        cleaned_word = word.strip(string.punctuation)
        if cleaned_word:
            frequency[cleaned_word] = frequency.get(cleaned_word, 0) + 1
    return frequency


# 8. Longest Word
def longest_word(text):
    words = text.split()
    longest = ""
    for word in words:
        cleaned_word = word.strip(string.punctuation)
        if len(cleaned_word) > len(longest):
            longest = cleaned_word
    return longest


# 9. Analyze Text
def analyze_text(text):
    print("\n=== TEXT ANALYSIS ===")

    total_words = count_words(text)
    total_vowels = count_vowels(text)
    total_consonants = count_consonants(text)
    reversed_version = reverse_text(text)
    palindrome_status = is_palindrome(text)
    text_without_vowels = remove_vowels(text)
    longest = longest_word(text)
    frequency = word_frequency(text)

    print(f"Words: {total_words}")
    print(f"Vowels: {total_vowels}")
    print(f"Consonants: {total_consonants}")
    print(f"Reversed: {reversed_version}")
    print(f"Palindrome: {'Yes' if palindrome_status else 'No'}")
    print(f"Without vowels: {text_without_vowels}")
    print(f"Longest word: {longest} ({len(longest)} letters)")

    print("Word Frequency:", end=" ")
    freq_output = []
    for word, count in frequency.items():
        freq_output.append(f"{word}: {count}")
    print(", ".join(freq_output))


# Main Execution
user_text = input("Enter text: ")
analyze_text(user_text)
```

```python
# Number System Functions

import math


# 1. Factorial
def factorial(n):
    if n < 0:
        return "Factorial not defined for negative numbers."
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result


# 2. Prime Check
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True


# 3. Fibonacci (nth term)
def fibonacci(n):
    if n <= 0:
        return "Position must be positive."
    if n == 1:
        return 0
    if n == 2:
        return 1

    a, b = 0, 1
    for _ in range(3, n + 1):
        a, b = b, a + b
    return b


# 4. Sum of Digits
def sum_of_digits(n):
    return sum(int(digit) for digit in str(abs(n)))


# 5. Reverse Number
def reverse_number(n):
    sign = -1 if n < 0 else 1
    reversed_num = int(str(abs(n))[::-1])
    return sign * reversed_num


# 6. Armstrong Number
def is_armstrong(n):
    digits = str(abs(n))
    power = len(digits)
    total = sum(int(digit) ** power for digit in digits)
    return total == abs(n)


# 7. GCD
def gcd(a, b):
    while b:
        a, b = b, a % b
    return abs(a)


# 8. LCM
def lcm(a, b):
    if a == 0 or b == 0:
        return 0
    return abs(a * b) // gcd(a, b)


# 9. Perfect Number
def is_perfect_number(n):
    if n <= 1:
        return False
    divisors_sum = 1
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            divisors_sum += i
            if i != n // i:
                divisors_sum += n // i
    return divisors_sum == n


# 10. Menu System
def math_menu():
    while True:
        print("\n====== NUMBER SYSTEM FUNCTIONS ======")
        print("1. Factorial")
        print("2. Prime Check")
        print("3. Fibonacci")
        print("4. Sum of Digits")
        print("5. Reverse Number")
        print("6. Armstrong Number")
        print("7. GCD")
        print("8. LCM")
        print("9. Perfect Number")
        print("10. Exit")

        choice = input("Enter choice (1-10): ")

        if choice == "10":
            print("Exiting program. Goodbye!")
            break

        try:
            if choice == "1":
                n = int(input("Enter number: "))
                print("Factorial:", factorial(n))

            elif choice == "2":
                n = int(input("Enter number: "))
                print("Is Prime" if is_prime(n) else "Not Prime")

            elif choice == "3":
                n = int(input("Enter position (n): "))
                print("Fibonacci:", fibonacci(n))

            elif choice == "4":
                n = int(input("Enter number: "))
                print("Sum of digits:", sum_of_digits(n))

            elif choice == "5":
                n = int(input("Enter number: "))
                print("Reversed number:", reverse_number(n))

            elif choice == "6":
                n = int(input("Enter number: "))
                print("Armstrong:" if is_armstrong(n) else "Not Armstrong")

            elif choice == "7":
                a = int(input("Enter first number: "))
                b = int(input("Enter second number: "))
                print("GCD:", gcd(a, b))

            elif choice == "8":
                a = int(input("Enter first number: "))
                b = int(input("Enter second number: "))
                print("LCM:", lcm(a, b))

            elif choice == "9":
                n = int(input("Enter number: "))
                print("Perfect Number:" if is_perfect_number(n) else "Not Perfect")

            else:
                print("Invalid choice. Please select 1-10.")

        except ValueError:
            print("Invalid input. Please enter integer values only.")


# Run the program
math_menu()
```
