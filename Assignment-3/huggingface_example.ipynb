```python
# huggingface_example.py
import os
import requests

# ---------------- API CONFIG ----------------
api_key = os.getenv("HUGGINGFACE_API_KEY")
if not api_key:
    print("Error: HUGGINGFACE_API_KEY not set")
    exit()

API_URL = "https://router.huggingface.co/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}

# ---------------- QUERY FUNCTION ----------------
def query_huggingface(prompt):
    try:
        payload = {
            "model": "meta-llama/Llama-3.1-8B-Instruct:cerebras",  # ✅ free + confirmed working
            "messages": [
                {"role": "user", "content": prompt}
            ],
            "max_tokens": 500,
            "temperature": 0.7
        }
        response = requests.post(API_URL, headers=headers, json=payload, timeout=60)

        if response.status_code == 503:
            return "Model is loading, please wait ~20 seconds and retry."
        if response.status_code != 200:
            return f"Error {response.status_code}: {response.text}"

        data = response.json()
        return data["choices"][0]["message"]["content"]

    except Exception as e:
        return f"Error: {str(e)}"

# ---------------- MAIN ----------------
if __name__ == "__main__":
    user_prompt = input("Enter your prompt: ")
    print("\nQuerying HuggingFace...\n")
    result = query_huggingface(user_prompt)
    print("Response:")
    print(result)
```
